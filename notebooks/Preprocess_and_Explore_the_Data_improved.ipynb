```python
# Import necessary libraries and scripts
import pandas as pd
import matplotlib.pyplot as plt  

from Preprocess_and_Explore_the_Data_improved import (
    load_and_clean_data, perform_eda, detect_outliers, save_data, 
    plot_outliers, interactive_plot
)

# Define stock tickers and date range
tickers = {"TSLA": "Tesla", "BND": "Vanguard Total Bond Market ETF", "SPY": "S&P 500 ETF"}
start_date = "2015-01-01"
end_date = "2024-10-31"
```

```python
# Block 1: Perform EDA
for ticker, name in tickers.items():
    print(f"Performing EDA for {name} ({ticker})...")
    data = load_and_clean_data(ticker, start_date, end_date)  
    figures = perform_eda(data, ticker)
    for fig in figures:
        plt.show()  # Show each plot one after another

```

```python
# Block 2: Detect and plot outliers
for ticker, name in tickers.items():
    print(f"Detecting and plotting outliers for {name} ({ticker})...")
    data = load_and_clean_data(ticker, start_date, end_date)  
    data = detect_outliers(data, column='Close')  
    plot_outliers(data, column='Close', ticker=ticker)
```

```python
# Block 3: Interactive visualization
for ticker, name in tickers.items():
    print(f"Creating interactive visualization for {name} ({ticker})...")
    data = load_and_clean_data(ticker, start_date, end_date)  
    interactive_plot(data, column='Close', ticker=ticker)

```

```python
# Block 4: Save processed data
combined_data = []  
for ticker, name in tickers.items():
    print(f"Saving data for {name} ({ticker})...")
    data = load_and_clean_data(ticker, start_date, end_date)  
    data = detect_outliers(data, column='Close')  
    save_data(data, filename=f"{ticker}_processed.csv")  
    combined_data.append(data)
combined_df = pd.concat(combined_data)
save_data(combined_df, "all_tickers_with_outliers.csv")
```
