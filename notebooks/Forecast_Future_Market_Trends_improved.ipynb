```python
import pandas as pd
import sys
sys.path.append('/home/bam/Desktop/week-12/scripts/')
from Forecast_Future_Market_Trends_improved import load_and_preprocess_data, arima_forecast, sarima_forecast, lstm_forecast, plot_forecast, save_forecast_data, save_forecast_analysis
import numpy as np
import warnings

warnings.filterwarnings("ignore")

# Define the stock tickers and names
tickers = {"TSLA": "Tesla", "BND": "Vanguard Total Bond Market ETF", "SPY": "S&P 500 ETF"}

# Loop over each ticker to perform forecasting and analysis
for ticker, name in tickers.items():
    # Load and preprocess data (from Task-2 structure)
    data = load_and_preprocess_data(ticker)
    
    # Forecast the next 6-12 months (252 trading days ~ 1 year)
    forecast_steps = 252  # Adjust according to how far ahead you want to forecast
    
    # ARIMA Forecast
    arima_forecast_data = arima_forecast(data, forecast_steps)
    
    # SARIMA Forecast
    sarima_forecast_data = sarima_forecast(data, forecast_steps)
    
    # LSTM Forecast
    lstm_model_path = f'{ticker}_lstm_model.h5'  # Path to pre-trained LSTM model
    lstm_forecast_data = lstm_forecast(data, forecast_steps, lstm_model_path)
    
    # Generate future dates for the forecast
    forecast_dates = pd.date_range(start=data.index[-1], periods=forecast_steps + 1, freq='B')[1:]
    
    # Plotting forecast for each model
    plot_forecast(data, forecast_dates, arima_forecast_data, sarima_forecast_data, lstm_forecast_data, ticker, name)
    
    # Save the forecasted data
    save_forecast_data(forecast_dates, arima_forecast_data, sarima_forecast_data, lstm_forecast_data, ticker)

    # Trend Analysis
    historical_mean = data['Close'].mean()
    arima_forecast_mean = np.mean(arima_forecast_data)
    sarima_forecast_mean = sarima_forecast_data.predicted_mean.mean()
    lstm_forecast_mean = np.mean(lstm_forecast_data)

    print(f"Historical Mean: {historical_mean}")
    print(f"ARIMA Forecast Mean (Next 6 Months): {arima_forecast_mean}")
    print(f"SARIMA Forecast Mean (Next 6 Months): {sarima_forecast_mean}")
    print(f"LSTM Forecast Mean (Next 6 Months): {lstm_forecast_mean}")

    # Identifying trends based on forecasted data
    trend = "Upward" if arima_forecast_mean > historical_mean else "Downward"
    print(f"Trend Analysis (ARIMA): The forecast suggests a {trend} trend.")

    trend = "Upward" if sarima_forecast_mean > historical_mean else "Downward"
    print(f"Trend Analysis (SARIMA): The forecast suggests a {trend} trend.")

    trend = "Upward" if lstm_forecast_mean > historical_mean else "Downward"
    print(f"Trend Analysis (LSTM): The forecast suggests a {trend} trend.")

    # Volatility and Risk Analysis for SARIMA forecast (confidence intervals)
    volatility_range = sarima_forecast_data.conf_int().iloc[:, 1] - sarima_forecast_data.conf_int().iloc[:, 0]
    max_volatility = volatility_range.max()
    min_volatility = volatility_range.min()

    print(f"Maximum Volatility (Uncertainty) in SARIMA Forecast: {max_volatility}")
    print(f"Minimum Volatility in SARIMA Forecast: {min_volatility}")

    # Market Opportunities and Risks for SARIMA forecast
    opportunity_risk = "Potential Market Opportunity" if sarima_forecast_mean > historical_mean else "Potential Market Risk"
    print(f"Market Opportunity/Risk Analysis (SARIMA): {opportunity_risk}")

    # Save forecast analysis to CSV
    save_forecast_analysis(forecast_dates, arima_forecast_data, sarima_forecast_data, lstm_forecast_data, volatility_range, trend, opportunity_risk, ticker)

print("Forecasting and Analysis Completed!")
```
